# Representation of one-hidden-layer neural network

+ The class of functions in approximation:

$$y \approx  \sum_{i=1}^m c_i f(\mathbf{x}; \mathbf{w}_i, b_i) $$

+ The (emperical) mean squared loss
$$L(\theta)= \sum_{j=1}^N \bigg | y_j -\sum_{i=1}^m c_i f(\mathbf{x}_j; \mathbf{w}_i, b_i) \bigg|^2 $$

  where all parameters are summarized as a variable $\theta= (c_i, \mathbf{w}_i, b_i)$, $i=1,\ldots, m$. $\mathbf{w}_i \in \mathbb{R}^n$.

+ There are  $m*(2+ n)$ unknowns.

## plot the input-output function 
$$ x \to \sum_{i=1}^m c_i f({x}; {w}_i, b_i) $$ for a given set of parameters $\theta= (c_i, \mathbf{w}_i, b_i)$, $i=1,\ldots, m$.

Readers may refer to further discussion
http://www.magnushansson.xyz/2017/10/17/ReLU-Nets-are-Splines.html

```python

import numpy as np
from matplotlib import pyplot as plt

# If it is run in jupyter notebook, please use the code below
%matplotlib inline

def my_relu(x,w,b,c,m):
    # x is a number
    # w,b,c are m-dim vectors
    y=0
    for i in range(m):
        y=y+c[i]*max(w[i]*x+b[i],0)
    return y

#This is the number of relu functions
m=1

#w and b are parameters of relu functions, they are m-dim vectors
w=np.array([-1,  1 ])
b=np.array([-0.5, -0.5])
#c is the weight of each relu function, it is also a m-dim vector
c=np.array([1.0, 1.0])

m=len(c)

n_plot=201;
x_show=np.linspace(-2,2,n_plot)
y_show=np.zeros(n_plot)
for ishow in range(n_plot):
    y_show[ishow]=my_relu(x_show[ishow],w,b,c,m)
    
fig1=plt.figure()
ax1=fig1.add_subplot(111)
p1_1=ax1.plot(x_show,y_show,'-b')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')


plt.show()
```

##   plot of  activation functions:  sigmoid, tanh, RELU, LeakyLU, exponential LU

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def sigmoid(x):
    return 1./(1+np.exp(-x)) 
def tanh(x):
    return  np.tanh(x)  
def ReLU(x):
    return  max(0,x) 
def LeakyReLU(x):
    return  max(0.1*x,x)
def ELU(x):
    alpha= 10  #try =1.0 =10
    if  x > 0:
        return x;
    else :
        return alpha*(np.exp(x)-1)
 
my_act =  {0 : sigmoid,
           1 : tanh,
           2 : ReLU,
           3 : LeakyReLU,
           4 : ELU,
}

x_show=np.linspace(-5,5,101)
y_show=np.zeros(101)
for ishow in range(101):
    y_show[ishow]=my_act[4](x_show[ishow])

fig1=plt.figure()
ax1=fig1.add_subplot(111)
p1_1=ax1.plot(x_show,y_show,'b')
plt.show() 
    
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
%matplotlib auto


Q=np.array([[1,0],[0,1]])
# Q=np.array([[1,0],[0,-1]])
# Q=np.array([[1,0],[1,0]])
# Q=np.array([[1,.5],[.5,0]])
# Q=np.array([[0,1],[1,0]])
# Q=np.array([[0,2],[-1,0]])


def check_symmetric(a, rtol=1e-05, atol=1e-08):
    return np.allclose(a, a.T, rtol=rtol, atol=atol)


def is_pos_def(x):
    return np.all(np.linalg.eigvals(x) > 0)

print('eigenvalues are', np.linalg.eigvals(Q))

if check_symmetric (Q):
    print('Q is symmetric')
    if is_pos_def(Q):
        print('Q is positive definite')
    else:
        print('Q is not positive definite')
else:
    print('Attention: Q is asymmetric')

    
b=np.array([0,0])
c=0
    
X=np.linspace(-1,1,121)
Y=np.linspace(-1,1,121)
x,y=np.meshgrid(X,Y)
z=0.5*(x*Q[0,0]*x+y*Q[1,0]*x+x*Q[0,1]*y+y*Q[1,1]*y)+b[0]*x+b[1]*y+c

fig1=plt.figure()
ax1=fig1.add_subplot(111,projection='3d')
p1_1=ax1.plot_surface(x,y,z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
fig1.colorbar(p1_1,shrink=0.5,aspect=5)

fig, ax = plt.subplots()
CS = ax.contour(x, y, z)
ax.clabel(CS,   inline=True,   fontsize=10)

plt.axis('square')

plt.show()
```

 #   Lasso regression  in 1D
 
 
 ## Example: exact solution of constraint Lasso in 1D
 Given $v\in \mathbb{R}$ and $\rho>0$, consider 
$$
\begin{array}{ll}
\min _{\beta} & |\beta-v|^2 \\
\text { subject to } & |\beta| \leq \rho
\end{array}
$$ The solution  is $$\beta^* =\begin{cases} v , &  |v|\leq \rho\\ \rho\operatorname{sign}(v),&  |v|\geq \rho\end{cases}$$ 

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib auto
import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
 

def conlasso(x,rho):
    if abs(x)>rho:
        y=rho*np.sign(x)
    else:
        y=x
    return y

rho=0.5



x=np.linspace(-1,1,21)
y=np.zeros(21)

for i in range(21):
    y[i]=conlasso(x[i],rho)
    
fig1=plt.figure()
ax1=fig1.add_subplot(111)
p1=ax1.plot(x,y,"--")
plt.xlabel(r'$v$',fontsize=32)
try: 
    plt.ylabel(r"$\beta^\star$ ",fontsize=32)
except: 
    pass

plt.title(r'optimal $\beta$ at $\rho=$'+str(rho),fontsize=22)


plt.show()
```

 ## Example: exact solution of  Lasso in penalty form
 Given $v\in \mathbb{R}$ and $\lambda >0$, consider 
$$
\begin{array}{ll}
\min _{\beta} & L(\beta):=\frac12(\beta-v)^2+ \lambda |\beta|  
\end{array}
$$ The solution   is the **soft thresholding operator**
## $$\beta^* =\mathcal{S}_{\lambda}(v)=\begin{cases} 0 , &  |v|\leq \lambda\\ v- \operatorname{sign}(v)\lambda,&  |v|\geq \lambda \end{cases}$$ 

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})

def  lasso_obj(x,v,lam):
    return 0.5*np.linalg.norm(x-v)**2 + lam*np.linalg.norm(x) 

def  soft_thrshld(v,lam):
    if (abs(v)< lam):
        return 0
    else: 
        return v - lam*np.sign(v)



v = 4 ; # try v = 5
lam = 3;

optimal_beta = soft_thrshld(v,lam) 
optimal_value = lasso_obj(optimal_beta,v,lam)

x=np.linspace(optimal_beta-5,optimal_beta+5,51)
obj=np.zeros(51)
for i in range(51):
    obj[i]=lasso_obj (x[i],v,lam)
    
fig1=plt.figure(figsize=(12,4))
ax1=fig1.add_subplot(121)
p1=ax1.plot(x,obj,"--")
ax1.set_title(r"The objective function  ($v=$" +str(v)+ ", $\lambda=$"+str(lam)+")",fontsize=22)
ax1.set_xlabel(r'$\beta$',fontsize=32)
ax1.set_ylabel(r'$L(\beta)$',fontsize=32)


p1=ax1.plot(optimal_beta,optimal_value,"o", color='green', markersize=12)


v_show = np.linspace(-10,10, 51)
beta_show = np.zeros( 51)
    
for i in range(51):
        beta_show[i] = soft_thrshld (v_show[i],lam)

ax2=fig1.add_subplot(122)
p2=ax2.plot(v_show,beta_show,linewidth=4)
ax2.set_title(r"soft-thresholding operator ($\lambda=$"+str(lam)+")",fontsize=22)
ax2.set_xlabel(r'$v$' ,fontsize=32)
ax2.set_ylabel(r'$\beta^\star$',fontsize=32)

plt.subplots_adjust(wspace=0.3)

plt.show()
 
```
